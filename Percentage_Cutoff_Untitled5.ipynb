```python
grp1 = data.groupby(['Customer_ID', 'Brand','Date'])

x=grp1.aggregate(np.sum)
x
```

```python
data1= x.groupby(['Customer_ID', 'Date',]).agg({"Brand_Count":(['min','max','sum'])}).reset_index()
print(data1)
```

```python
data2= x.groupby(['Date']).agg(Brand_Count_max=('Brand_Count', 'max'),Brand_Count_sum=('Brand_Count', 'sum')).reset_index()
data2
```

```python
data2['Percentage cutoff %']= data2['Brand_Count_max']/data2['Brand_Count_sum']*100
data2
```
